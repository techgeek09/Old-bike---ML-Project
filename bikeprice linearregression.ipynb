```python
lnr = LinearRegression()  
```

```python
lnr.fit(x_train,y_train) 
```

```python
# Assuming 'Owner' is the column with categorical data
x_train = pd.get_dummies(x_train, columns=['owner'])
```

```python
lnr.score(x_train,y_train)*100   # 75% scored 
```

```python
# unseen data or test data or validation data (x_test,y_test)
lnr.score(x_test,y_test)*100 
```
